### Visualizations and matplotlib

When working with data, it is often inconvenient to draw any conclusions when looking at a table and numbers in particular, so it is important to be able to visualize data. Here the key rules for the design of schedules for **all** homework will be described.

matplotlib, of course, has [documentation](https://matplotlib.org/users/index.html) with a lot of [examples](https://matplotlib.org/examples/), but to get started it is enough to know about several basic types of charts:
- plot - a regular point-by-point plot, which can be used to represent curves or individual points;
- hist - a histogram showing the distribution of a certain value;
- scatter - a graph showing the relationship between two values;
- bar — a bar chart showing the relationship between a quantitative value and a categorical one.

In this task, you will try to build one of them. Do not forget about the basic principles of building decent charts:
- axes must be signed, and not too small;
- the chart must have a title;
- if several graphs are shown, then an explanatory legend is required;
- all lines on the charts should be clearly visible (there are no similar colors or colors merging with the background);
- if a value is displayed that has an obvious range of values ​​(for example, percentages can be from 0 to 100), then it is desirable to scale the axis for the entire range of values ​​(the exception is the case when you need to show a small difference that is imperceptible in such scales);
- the grid on the chart helps to estimate the values ​​in points by eye, this is usually useful, so it's better to draw it;
- if the distribution on the histogram has heavy tails, it is better to use a logarithmic scale.

### 2. Distribution of students by courses.

Now you need to assign students to fall elective courses based on their preferences.

Algorithm for the distribution of students by courses:
1. By default, 1 group (up to 30 students) can study in each discipline of choice for 3 and 4 courses. Exceptions are described below. In blended disciplines for fourth-year students, the number of places is not limited.
2. The first wave of selection is carried out. For each discipline, a list of those who indicated it as the first priority is formed (if a student has to choose two elective courses, then for him the disciplines that he indicated as the first and second priority are considered as the disciplines of the first priority). If there are more applicants than places, then the top percentile of the rating is selected.
3. In the disciplines where there are places left after the first wave, lists of those who chose them as the second priority are formed, and the places are filled by the best students in terms of the rating percentile. After that, the same procedure is carried out for disciplines of the third priority.
4. If the student did not get the required number of courses based on the results of three waves, the study office contacts him and resolves the issue on an individual basis.

We draw your attention to the following details:

- By default, a student chooses one fall and one spring elective course, and fourth-year students choose one blended course. Students of groups of the 17th year of MOS and TI specializations choose 2 autumn and 2 spring courses of their choice, also students of groups of 18' specializations of MI choose 2 autumn courses.

- MEP students cannot take the Spring Elective Course in Machine Learning 2.

- Don't touch blended courses, they don't need to be distributed among them, they don't affect other courses in any way.
